```r
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
library(ggplot2)
library(cowplot)
```

```r
data <- read_csv("abc/vaccination-administration-2.csv")

tail(data)
```

```r
data_canada<- data|>
filter(prename=="Canada")
data_canada<-tail(data_canada, 20)
data_canada
```

```r
data_canada<- data_canada|>
select(report_date, numtotal_all_administered, as_of_date)
data_canada
```

# Cumulative dministrations (x-axis starts: 2023-01-12)  

```r
options(repr.plot.width = 12, repr.plot.height = 7)
ggplot(data_canada, aes(x=report_date,y=numtotal_all_administered))+
geom_point()+
geom_line()+
ggtitle("at least 1 dose")+
labs(x="Report Date",y="Vaccine Administered (All Doses)")
```
